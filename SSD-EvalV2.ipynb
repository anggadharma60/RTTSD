<h1>Function</h1>

<h2>Intersection of Union (IoU)</h2>

```python
def compute_iou(GT, P):
    inter_xmin = max(GT[0], P[0]) #get maximum xmin
    inter_ymin = max(GT[1], P[1]) #get maximum ymin
    inter_xmax = min(GT[2], P[2]) #get minimum xmax
    inter_ymax = min(GT[3], P[3]) #get minimum ymax

    #get length
    inter_w = max(inter_xmax - inter_xmin, 0)
    inter_h = max(inter_ymax - inter_ymin, 0)
    inter_area = inter_w * inter_h
#     print("inter_xmin, inter_ymin, inter_xmax, inter_ymax, inter_w, inter_h, inter_area")
#     print( inter_xmin, inter_ymin ,inter_xmax , inter_ymax, inter_w, inter_h, inter_area)
    if (inter_area == 0):
#         print("There is no intersection area")
        return 0
    else:
        GT_area = (GT[2]-GT[0])*(GT[3]-GT[1])
        P_area = (P[2]-P[0])*(P[3]-P[1])
        union_area = GT_area + P_area - inter_area
#         print("GT_area, P_area, union_area")
#         print(GT_area, P_area, union_area)
        iou = float(inter_area / union_area)
#         print(iou)

        return float(format(iou, '.2f'))
```

<h2>Filter</h2>

```python
def filter_bb(arrGT, boxes, classes, scores, count, labels,imW, imH, threshold):
    results = []
    keep = []
    Box = []
    Class = []
    Score = []
    for i in range(count):
        ymin = int(max(1,(boxes[i][0] * imH)))
        xmin = int(max(1,(boxes[i][1] * imW)))
        ymax = int(min(imH,(boxes[i][2] * imH)))
        xmax = int(min(imW,(boxes[i][3] * imW)))
        
#         xmin, xmax, ymin , ymax, classid, classname
        arrP= [xmin, ymin, xmax, ymax]
        iou = compute_iou(arrGT, arrP)
        iou = float(format(iou, '.2f'))
        if iou>= threshold:
            classid = int(classes[i]+1)
            score = float(format(scores[i], '.2f'))
            object_name = labels[classid-1]
            Detected = 'Positive'
            if classid == arrGT[4]:
                Classified = 'True'
            else:
                Classified = 'False'
                
            arrP.append(classid)
            arrP.append(object_name)
            arrP.append(score)
            arrP.append(iou)
            arrP.append(Detected)
            arrP.append(Classified)
            results.append(arrP)
            
#             print("Match BB on GT")
            break
        
        else:
            pass
#             print("Iou under threshold")
   
    return results
```

```python
label = 'TFLite_label.txt'
PATH_TO_TFLITE_LABEL = os.path.join('data', label)

def load_tflite_label_map(path):
    """Loads the labels file. Supports files with or without index numbers."""
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        ret = {}
        for row_number, content in enumerate(lines):
            pair = re.split(r'[:\s]+', content.strip(), maxsplit=1)
        
            if len(pair) == 2 and pair[0].strip().isdigit():
                ret[int(pair[0])] = pair[1].strip()
          
            else:
                ret[row_number] = pair[0].strip()
            # print(pair[0].strip())
          
    return ret
print(load_tflite_label_map(PATH_TO_TFLITE_LABEL))

NUM_CLASSES = len(load_tflite_label_map(PATH_TO_TFLITE_LABEL))
print(NUM_CLASSES)
```

```python
import platform
import importlib
EDGETPU_SHARED_LIB = {
  'Linux': 'libedgetpu.so.1',
  'Windows': 'edgetpu.dll'
}[platform.system()]

   
def build_interpreter(result, USE_TPU):
    platform = check_platform()
    pkg = importlib.util.find_spec('tflite_runtime')
    # pkg = importlib.util.find_spec('tensorflow')
    if pkg:
        from tflite_runtime.interpreter import Interpreter
        if USE_TPU:
            from tflite_runtime.interpreter import load_delegate
    else:
        from tensorflow.lite.python.interpreter import Interpreter
        if USE_TPU:
            from tensorflow.lite.python.interpreter import load_delegate
    if USE_TPU:
        if (platform == "Linux"):
            interpreter = Interpreter(model_path=result[1], num_threads=4,
                                      experimental_delegates=[load_delegate('libedgetpu.so.1')])
    
        if (platform == "Windows"):
            interpreter = Interpreter(model_path=result[1], num_threads=4,
                                      experimental_delegates=[load_delegate('edgetpu.dll')])
    else:
        interpreter = Interpreter(model_path=result[0], num_threads=4)
    
    return interpreter

def searchFile(folder):
    result = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(".tflite"):
                result.append(os.path.join(os.getcwd(), root, file))

    return result

def check_platform():
    return platform.system()
# result = searchFile(output_directory_tflite1)
# print(result)
# build_interpreter(result, 0)
# print(build_interpreter(False, MODEL_GRAPH))
```

```python
import pandas as  pd
subcase = [1,2,3,4,5,6,7,8,9]


def confusion_matrix_save(csv_save_dir):
    
    test = read_csv(csv_save_dir)
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    times = 0
    for i in range(len(test)):

        if (test[i][16] == True ):
            if (test[i][15] == "Positive" ):
                TP+=1
            elif (test[i][15] == "Negative" ):
                TN+=1
        elif (test[i][16] == False ):
            if (test[i][15] == "Positive" ):
                FP+=1   
            elif (test[i][15] == "Negative" ):
                FN+=1
        times+=test[i][17]
    accuracy = float((TP+TN)/(TP+TN+FP+FN)) if ((TP+TN+FP+FN) != 0) else  0
    accuracy = float(format(accuracy, '.2f'))
    precision = float(TP/(TP+FP)) if ((TP+FP) != 0) else 0
    precision = float(format(precision, '.2f'))
    recall = float(TP/(TP+FN)) if ((TP+FN) != 0) else  0
    recall = float(format(recall, '.2f'))
    avg_times = times/len(test)
    avg_times = float(format(avg_times, '.3f'))
#     sensivity = float(TP/(TP+FN)) if ((TP+FN) != 0) else  0
#     specificity = float(TN/(TN+FP)) if ((TN+FP) != 0) else  0
#     print("True Positive:", TP)
#     print("True Negative:", TN)
#     print("False Positive:", FP)
#     print("False Negative:", FN)
#     print("Accuracy:", accuracy)
#     print("Precision:", precision)
#     print("Recall:", recall)

    return {'file':csv_save_dir, 'TP':TP, 'FP': FP, 'FN':FN, 'TN':TN, 'acc':accuracy, 'prec':precision, 'rec':recall, 'avg_times': avg_times}

    
def eval_matrix(subcase):

    result = []
    for i in range(len(subcase)):
        case = str(subcase[i])
        iteration1 = '50l'
        iteration2= '100l'
        iteration3= '150l'
        folder = 'SSD'
        PATH_TO_TEST_IMAGES_DIR = 'test/300l/'
      

        path_eval_tflite1 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration1+'/'+'0'+'/'
        path_eval_tflite2 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration2+'/'+'0'+'/'
        path_eval_tflite3 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration3+'/'+'0'+'/'
        path_eval_tflite4 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration1+'/'+'1'+'/'
        path_eval_tflite5 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration2+'/'+'1'+'/'
        path_eval_tflite6 =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+case+'/'+iteration3+'/'+'1'+'/'

        

        csv_eval_tflite1 = path_eval_tflite1+'Result_tflite_'+iteration1+'.csv'
        csv_eval_tflite2 = path_eval_tflite2+'Result_tflite_'+iteration2+'.csv'
        csv_eval_tflite3 = path_eval_tflite3+'Result_tflite_'+iteration3+'.csv'
        csv_eval_tflite4 = path_eval_tflite4+'Result_tflite_'+iteration1+'.csv'
        csv_eval_tflite5 = path_eval_tflite5+'Result_tflite_'+iteration2+'.csv'
        csv_eval_tflite6 = path_eval_tflite6+'Result_tflite_'+iteration3+'.csv'
#         confusion_matrix(csv_eval1)
#         confusion_matrix(csv_eval2)
        result.append(confusion_matrix_save(csv_eval_tflite1))
        result.append(confusion_matrix_save(csv_eval_tflite2))
        result.append(confusion_matrix_save(csv_eval_tflite3))
        result.append(confusion_matrix_save(csv_eval_tflite4))
        result.append(confusion_matrix_save(csv_eval_tflite5))
        result.append(confusion_matrix_save(csv_eval_tflite6))


    return result

def result_to_excel(result):
    
    PATH_TO_TEST_IMAGES_DIR = 'test/300l/'
    folder = 'SSD'

    path_excel =  PATH_TO_TEST_IMAGES_DIR+folder+'/'+'eval_excel.csv'
    

    df = pd.DataFrame(result)

    df.to_csv(path_excel, index = False, sep=';')

def result_to_txt(result):
    text = ''
    acc = 0
    index = 0
    for i in range(len(result)):
        
        
        if result[i]['acc'] > acc:
            acc = result[i]['acc']
            index = i
        
        text+="==========================================================================\n"
        text+= 'File: %s' % result[i]['file']+'\n'
        text+="--------------------------------------------------------------------------\n" 
        if i%2==1:
            text+="|                              With TPU                                  |\n"   
        if i%2==0:
            text+="|                             Without TPU                                |\n"    
        text+="--------------------------------------------------------------------------\n" 
        text+= 'TP: %d\t FN: %d' % (result[i]['TP'], result[i]['FN'])+'\n'
        text+= 'FP: %d\t TN: %d' % (result[i]['FP'], result[i]['TN'])+'\n'
        text+= 'Accuracy  : %.2f' % result[i]['acc']+'\n'
        text+= 'Precision : %.2f' % result[i]['prec']+'\n'
        text+= 'Recall    : %.2f' % result[i]['rec']+'\n'
        text+= 'Avg Times : %.3f' % result[i]['avg_times']+'s\n'
        text+="==========================================================================\n"
    print(text)
    print("Best Accuracy:")
    print(result[index])

def eval_summary(result):
    acc = 0
    index = 0
    print('Subcase'+'   TP'+'   FN'+'   FP'+'   TN'+'   Accuracy'+'   Precision'+'   Recall'+'   Avg Times')
    for i in range(len(result)):
        if result[i]['acc'] > acc:
            acc = result[i]['acc']
            index = i
            
        temp = result[i]['file'].split("/")
        subcase = temp[3]+'/'+temp[4]+'/'+temp[5]+'   '
        print(subcase+'%d   %d   %d   %d   %.2f   %.2f   %.2f   %.3fs' % (result[i]['TP'], result[i]['FN'], result[i]['FP'], result[i]['TN'], result[i]['acc'],result[i]['prec'], result[i]['rec'], result[i]['avg_times']))
    print("Best Accuracy:")
    print(result[index])
    
result = eval_matrix(subcase)
# print(result)
```
